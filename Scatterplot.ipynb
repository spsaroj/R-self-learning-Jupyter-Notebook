```r
library(datasets)
head(mtcars)  #mtcars is the dataset
```

Since we are doing the scatterplot for wt and mpg, let's see the those separately.

```r
hist(mtcars$wt)
hist(mtcars$mpg)
```

Since we are giving two quantitative variables, R can recognize it and will make a scatterplot.

```r
plot(mtcars$wt, mtcars$mpg)  #Normal plot
```

```r
plot(mtcars$wt, mtcars$mpg,
  pch = 19,         # Solid circle
  cex = 1.5,        # Make 150% size
  col = "#cc0000",  # Red
  main = "MPG as a Function of Weight of Cars",
  xlab = "Weight (in 1000 pounds)",
  ylab = "MPG")
```
