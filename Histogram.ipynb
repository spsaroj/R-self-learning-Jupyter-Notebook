## Histogram by groups

In our database we have 3 different species of Iris. So, to see them side by side, we use this approach.
This approach is known as small multiples.

par(mfrow = c(3, 1))      #Put graphs in 3 rows and 1 column
where par means parameter, c is concatinating (treat these 2 numbers as a single unit.), 

```r
par(mfrow = c(3, 1))
```

The thing in the [] is selector. selects only setosa in this case. xlim is limit for the x. breaks means how many bars we want in the histogram. main is the title.

```r
hist(iris$Petal.Width [iris$Species == "setosa"],
  xlim = c(0, 3),
  breaks = 9,
  main = "Petal Width for Setosa",
  xlab = "",
  col = "red")

hist(iris$Petal.Width [iris$Species == "versicolor"],
  xlim = c(0, 3),
  breaks = 9,
  main = "Petal Width for Versicolor",
  xlab = "",
  col = "purple")

hist(iris$Petal.Width [iris$Species == "virginica"],
  xlim = c(0, 3),
  breaks = 9,
  main = "Petal Width for Virginica",
  xlab = "",
  col = "blue")
```
