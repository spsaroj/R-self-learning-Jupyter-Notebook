```r
library(datasets)
head(iris)
```

```r
hist(iris$Petal.Length)
```

```r
summary(iris$Petal.Length)
summary(iris$Species)
```

```r
hist(iris$Petal.Length[iris$Species == "versicolor"], main = "Petal Length: Versicolor")
```

```r
# Virginica
hist(iris$Petal.Length[iris$Species == "virginica"],
  main = "Petal Length: Virginica")

# Setosa
hist(iris$Petal.Length[iris$Species == "setosa"],
  main = "Petal Length: Setosa")
```

```r
hist(iris$Petal.Length[iris$Petal.Length < 2], main = "Petal Length greater than 2")
```

```r
hist(iris$Petal.Length[iris$Species == "virginica" & iris$Petal.Length < 5.5], main = "Virginica Petal Length less than 5.5")
```

#### If we are going to use the same subsample for the many times,

i.setosa <- iris[iris$Species == "setosa", ]

```r
summary(i.setosa)
hist(i.setosa$Petal.Length)
```
