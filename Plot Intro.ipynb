```r
library(datasets)
head(iris)
```

```r
plot(iris$Species)  #This means use dataset iris and use variable Species

plot(iris$Petal.Length)
```

```r
plot(iris$Species, iris$Petal.Width)  # This is the plot between categorical and quantitative variable
```

Since boxplot is the most common plot for quantitative variable, it uses boxplot.

```r
plot(iris$Petal.Length, iris$Petal.Width)  # Plot for Quantitative variables only

plot(iris)   #plots the entire dataset
```

It is possible for us to change the way plots look which is demonstrated below.
pch means point character
xlab, ylab means x/y labels.

```r
plot(iris$Petal.Length, iris$Petal.Width,
  col = "#cc0000",  # Red
  pch = 19,
  main = "Iris: Petal Length vs. Petal Width",
  xlab = "Petal Length",
  ylab = "Petal Width")
```

```r
#Demonstration for plotting the formulas
plot(exp, 1, 5)
```

```r
#Demonstration for plotting the formulas
plot(dnorm, -3, +3,
  col = "#cc0000",
  lwd = 5,
  main = "Standard Normal Distribution",
  xlab = "z-scores",
  ylab = "Density")
```
